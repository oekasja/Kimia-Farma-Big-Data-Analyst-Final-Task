```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_product = pd.read_csv('kf_product.csv')
df_kantor_cabang = pd.read_csv('kf_kantor_cabang.csv')

# 1. Product Analysis
# Calculate descriptive statistics for the `price` column
price_stats = df_product['price'].describe().round(2)

# Print descriptive statistics with labels
print("\nProduct Price Statistics:")
print(f"Mean Price: {price_stats['mean']}")
print(f"Median Price: {price_stats['50%']}")
print(f"Min Price: {price_stats['min']}")
print(f"Max Price: {price_stats['max']}")
print(f"Standard Deviation of Price: {price_stats['std']}")

# Calculate the number of unique values in `product_category`
num_unique_categories = df_product['product_category'].nunique()
print(f"\nNumber of Unique Product Categories: {num_unique_categories}")

# If the number of unique values is 25 or less, print the frequency count of each unique value
if num_unique_categories <= 25:
    category_counts = df_product['product_category'].value_counts()
    print("\nFrequency Count of Product Categories:")
    print(category_counts.to_markdown(numalign="left", stralign="left"))

# 2. Branch Analysis
# Calculate descriptive statistics for the `rating` column
rating_stats = df_kantor_cabang['rating'].describe().round(2)

# Print descriptive statistics with labels
print("\nBranch Rating Statistics:")
print(f"Mean Rating: {rating_stats['mean']}")
print(f"Median Rating: {rating_stats['50%']}")
print(f"Min Rating: {rating_stats['min']}")
print(f"Max Rating: {rating_stats['max']}")
print(f"Standard Deviation of Rating: {rating_stats['std']}")

# Calculate the number of unique values in `branch_category`
num_unique_branch_categories = df_kantor_cabang['branch_category'].nunique()
print(f"\nNumber of Unique Branch Categories: {num_unique_branch_categories}")

# If the number of unique values is 25 or less, print the frequency count of each unique value
if num_unique_branch_categories <= 25:
    branch_category_counts = df_kantor_cabang['branch_category'].value_counts()
    print("\nFrequency Count of Branch Categories:")
    print(branch_category_counts.to_markdown(numalign="left", stralign="left"))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram for Product Prices
plt.figure(figsize=(10, 6))
sns.histplot(df_product['price'], bins=10, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Product Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# 2. Histogram for Branch Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df_kantor_cabang['rating'], bins=10, kde=True)
plt.title('Distribution of Branch Ratings')
plt.xlabel('Branch Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# 3. Bar Chart for Product Categories
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.title('Frequency of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# 4. Bar Chart for Branch Categories
plt.figure(figsize=(10, 6))
branch_category_counts.plot(kind='bar')
plt.title('Frequency of Branch Categories')
plt.xlabel('Branch Category')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # No rotation for branch categories
plt.show()
```
